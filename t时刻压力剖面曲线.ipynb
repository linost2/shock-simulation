```python
import matplotlib.pyplot as plt
import pandas as pd
path = r"D:\Youlin Zhu\2021实验室\分子动力学\数据处理\strainrate\sample100"
outfig = "P_100_t.tiff"
```

```python
df = pd.read_excel(path+"/"+"cut.xlsx",sheet_name=0, header=0,index_col=0)
```

```python
goal_xaxis="Coord1"
goal_yaxis="v_pressurexx"
plt.figure(num=1)
plt.rcParams['xtick.direction'] = 'in'
plt.rcParams['ytick.direction'] = 'in'
```

```python
T=[41,41.6,42.4,43,43.8]#感兴趣的时刻列表
color=["red","blue","green","black","orange","yellow","cyan","purple","pink",]
ci=0
```

```python
for i in T:
    xaxis=df.loc[i,goal_xaxis].values
    yaxis=df.loc[i,goal_yaxis].values
    plt.plot(xaxis,yaxis,color=color[ci],linewidth=1.5)
    ci+=1
    
plt.xticks(fontproperties = 'Times New Roman',size = 14)
plt.xlim(500,1750)
plt.ylim(-200000,200000)
plt.yticks([-100000,0,100000,200000,300000],[-10,0,10,20,30],fontproperties = 'Times New Roman',size = 14)
plt.xlabel("Position (nm)",fontdict={'family' : 'Times New Roman', 'size'   : 16})
plt.ylabel("Pressure_xx (GPa)",fontdict={'family' : 'Times New Roman', 'size'   : 16})
```
